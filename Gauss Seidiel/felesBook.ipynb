```python
import numpy as np
import math
```

```python
matrizA = np.array([[ 4, -1,  0,  0,  0], [-1,  4, -1,  0,  0],[ 0, -1,  4, -1,  0],[ 0,  0, -1,  4, -1],[ 0,  0,  0, -1,  4]])
vectorB = np.array([[ 1.],[-math.pi],[0.],[math.pi],[0.]])
vectorCerosAntes = np.array([0] * 5)
```

```python
iteracion = 0
while True:
    print("polo", iteracion)
    iteracion += 1
    xNueva = vectorCerosAntes.copy()
    for i in range(5):
        # Según yo ambas sumatorias van a dar un número negativo porque se multiplica en el primer caso -ai* xj elveado a la k, para k para darría positivos así que el producto da negativo en la primera sumatoria. En la segunda ya es negativa
        vectorCerosAntes[i] = (1 - 1.0518) * xNueva[i] + (1.0518 / matrizA[i, i])*((-1*np.dot(matrizA[i, :i], vectorCerosAntes[:i])) - np.dot(matrizA[i, i + 1:], xNueva[i + 1:] ) + vectorB[i])
    print(vectorCerosAntes)
    if (np.linalg.norm(xNueva - vectorCerosAntes)/np.linalg.norm(vectorCerosAntes)) < 1e-4: 
        print("se acabó") 
        break
```
